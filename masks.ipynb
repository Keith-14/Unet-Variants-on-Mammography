```python
import os
import pandas as pd
import shutil

BASE_DIR = r"C:\Users\KEITH FERNANDES\OneDrive\Desktop\MAMMOGRAPHY\CBIS-DDSM"

JPEG_ROOT = os.path.join(BASE_DIR, "jpeg")
MASKS_ROOT = os.path.join(BASE_DIR, "masks")
OUT_CSV = os.path.join(BASE_DIR, "segmentation_pairs.csv")

print("JPEG_ROOT:", JPEG_ROOT)
print("Exists?", os.path.exists(JPEG_ROOT))

os.makedirs(MASKS_ROOT, exist_ok=True)

pairs = []

for uid in os.listdir(JPEG_ROOT):
    uid_dir = os.path.join(JPEG_ROOT, uid)
    if not os.path.isdir(uid_dir):
        continue

    img_path = None
    mask_path = None

    for f in os.listdir(uid_dir):
        if not f.lower().endswith(".jpg"):
            continue

        if f.startswith("1-") and img_path is None:
            img_path = os.path.join(uid_dir, f)

        elif f.startswith("2-") and mask_path is None:
            mask_path = os.path.join(uid_dir, f)

    if img_path and mask_path:
        mask_dst = os.path.join(MASKS_ROOT, f"{uid}_mask.jpg")
        if not os.path.exists(mask_dst):
            shutil.copy(mask_path, mask_dst)

        pairs.append({
            "uid": uid,
            "image_path": img_path,
            "mask_path": mask_dst
        })

df_pairs = pd.DataFrame(pairs)
df_pairs.to_csv(OUT_CSV, index=False)

print("Total valid segmentation pairs:", len(df_pairs))
print("Saved to:", OUT_CSV)
```
